```r
library(ggplot2)
```

```r
ggplot(mtcars, aes(x=disp, y=mpg),main='mpg vs.disp') + geom_line() + geom_point()
```

```r
ggplot(mtcars, aes(factor(cyl))) + geom_bar()
```

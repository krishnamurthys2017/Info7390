```python
import pandas as pd
from pandas import *
```

```python
s1=DataFrame([[1,2,3],[7,0,-1],[0,2,4],[5,2,3],[2,4,1]],columns=['a','b','c'])
print(s1)
print("corr(a,b)=" , s1.a.corr(s1.b))
print ("corr(b,a)=" , s1.b.corr(s1.a))
print ("corr(a,c)=" , s1.a.corr(s1.c))
print ("corr(b,c)=" , s1.b.corr(s1.c))
print ("correlation matrix:")
print (s1.corr())
```
